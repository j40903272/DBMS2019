```python
import os
import numpy as np
import pandas as pd
from neo4jrestclient.client import GraphDatabase
```

```python
gdb = GraphDatabase("http://localhost:7474/", username="neo4j", password="toor")
df = pd.read_csv('EECS5178_dataset.csv')
df = df.where((pd.notnull(df)), None)
df.columns = ['ID', 'identity', 'department', 'grade', 'studentID', 'name', 'EN_name', 'phone', 'email', 'group']
```

```python
people = gdb.labels.create("Person")

for i in range(len(df)):
    n = gdb.nodes.create()
    for key, attr in zip(df.columns, df.iloc[i]):
        if isinstance(attr, np.integer):
            attr = int(attr)
        n[key] = attr
    people.add(n)
```

```python
from neo4jrestclient import client
my_node = gdb.query("MATCH (n:Person) WHERE n.studentID = 'r08946007' RETURN n", returns=(client.Node))[0][0]
peer_results = gdb.query("MATCH (n:Person) WHERE (n.department='資料科學學程碩士班' AND n.grade = ' 一年級') RETURN n;", returns=(client.Node))
peer_results = gdb.query("MATCH (n:Person) WHERE n.department='資料科學學程碩士班' RETURN n;", returns=(client.Node))
```

```python
for peer in peer_results:
    peer_node = peer[0]
    if peer_node != my_node:
        my_node.relationships.create("peer", peer_node)
        peer_node.relationships.create("peer", my_node)
```
