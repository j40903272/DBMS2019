```python
import os
import numpy as np
import pandas as pd
from neo4jrestclient.client import GraphDatabase
```

```python
gdb = GraphDatabase("http://localhost:7474/", username="neo4j", password="toor")
gdb.query("MATCH (n) DETACH DELETE n;")
df = pd.read_csv('108-1SQL_hobby_dataset.csv')
df = df.where((pd.notnull(df)), None)
df.columns = ['studentID', 'name'] + list(df.columns[2:])
df
```

```python
### print answer

# find hobby friend
dada_hobby = set(df[df.studentID == 'r08946007'].values[0][2:])
hobby_friends = []
hobby_friends_hobby = set()

for i in range(len(df)):
    hobby = set([j for j in df.iloc[i].values[2:] if j])
    if (dada_hobby & hobby):
        hobby_friends.append(df.name.iloc[i])
        hobby_friends_hobby |= set(hobby)
        continue

print(len(hobby_friends), hobby_friends)

# find foaf
new_hobby = set()
foaf = list()
for i in range(len(df)):
    if df.name.iloc[i] == '蔡O達':
        continue
    hobby = set([j for j in df.iloc[i].values[2:] if j])
    if (hobby_friends_hobby & hobby):
        foaf.append(df.name.iloc[i])
        new_hobby |= set(hobby)
        continue

new_hobby -= dada_hobby
print('new_hobby', len(new_hobby))
print('foaf', len(foaf))

# check
assert not (dada_hobby & new_hobby)
assert '蔡O達' not in foaf
```

```python
### use labels
gdb.query("MATCH (n) DETACH DELETE n;")

for i in range(len(df)):
    n = gdb.nodes.create()
    n['name'] = df.name.iloc[i]
    n['studentID'] = df.studentID.iloc[i]
    for hobby in df.iloc[i].values[2:]:
        if hobby:
            n.labels.add(hobby)

query = """MATCH (n {studentID:'r08946007'}), (m) 
           WHERE ANY(x IN labels(n) where x IN labels(m))
           RETURN DISTINCT m.name"""
results = gdb.query(query)
print('hobby friends', len(results))



query = """match (n{studentID:'r08946007'}), (m)
           where any(x in labels(n) where x in labels(m))
           with m, n
           match (m), (o)
           where any(x in labels(m) where x in labels(o)) and (o.studentID <> 'r08946007')
           return DISTINCT o.name, filter(y in labels(o) where not y in labels(n))"""

results1 = gdb.query(query)
print('foaf', len(results1))



query = """match (n{studentID:'r08946007'}), (m)
           where any(x in labels(n) where x in labels(m))
           with m, n
           match (m), (o)
           where any(x in labels(m) where x in labels(o)) and (o.studentID <> 'r08946007')
           unwind filter(y in labels(o) where not y in labels(n)) as new_hobby
           return distinct new_hobby"""

results2 = gdb.query(query)
print('new hobbies', len(results2))
```
